```python
# 연습 2
# - 각 정수를 중심으로 +/- tau의 폭으로 1의 값을 갖는 구형파열을 만들고 그래프를 그리시오
#
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def createRectPulse(tau):
    x = np.linspace(0,5,500)
    y = ((tau/2)>=np.abs((np.round(x)-x)))
    return x,y
```

```python
x, y = createRectPulse(0.3)
plt.figure(figsize=(20,5))
plt.plot(x,y)
plt.grid()
plt.xticks(np.arange(0,5.1,0.1))
plt.show()
```
