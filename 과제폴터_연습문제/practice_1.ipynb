```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getCircle(x1,y1,r):
    #d = np.linspace(0,360,2000)
    #x = r*np.cos(d*np.pi/180)+x1
    #y =r*np.sin(d*np.pi/180)+y1
    d = np.linspace(0,2*np.pi,200)
    x = r*np.cos(d)+x1
    y = r*np.sin(d)+y1
    return x,y,d
```

```python
x1 = 5;y1=10;r = 5

x, y,d = getCircle(x1,y1,r)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1);plt.grid();plt.title('x axis')
plt.plot(d,x,'r')
plt.xlabel('degree');plt.ylabel('x')
plt.subplot(1,3,2);plt.grid();plt.title('y axis')
plt.plot(d,y,'b')
plt.xlabel('degree');plt.ylabel('y');

plt.subplot(1,3,3);plt.grid();plt.title('Circle')
plt.plot(x,y,'g')
plt.xlabel('x');plt.ylabel('y')
plt.xticks(np.arange(x1-(r*1.2), x1+(r*1.2), 1));
plt.yticks(np.arange(y1-(r*1.2), y1+(r*1.2), 1));
```
