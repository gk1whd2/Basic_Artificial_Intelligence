```python
# 연습 3
# - 연습2에서 만든 구형파열을 이용해 값이 1일때는 위상이 pi, 0일때는 위상이 0인 정형파를 그리시오.
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def createRectPulse(tau,xLen=2):
    x = np.linspace(0,xLen,xLen*200)
    y = ((tau/2)>=np.abs(np.round(x)-x))
    return x,y
```

```python
def practice3(tau,f):
    rectX,rectY = createRectPulse(tau)
    y = np.cos(2*np.pi*f*rectX + (rectY)*np.pi)
    return rectX,y,rectY
```

```python
tau = 0.3
fre = 5
x, y,rectY = practice3(tau,fre)  # 입력 : tau = 0.5, frequency = 5

plt.figure(figsize=(15,5))
plt.xticks(np.arange(0,2,tau/2))
plt.yticks([-1,0,1])
plt.plot(x,y,'b')
plt.plot(x,rectY,'r--')
plt.grid()
```
