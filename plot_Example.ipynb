```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f1(x):
    return (x-2)*x*(x+2)

def f2(x):
    return 3*(x-1)*(x-2)

x = np.linspace(-3,3,20)

y1 = f1(x)
y2 = f2(x)

print(y1)
print(y2)

plt.subplot(1,2,1)
plt.plot(x,y1,'-r',label='f1')
plt.legend(), plt.grid()
plt.title('My first Python graphs')

plt.subplot(1,2,2)
plt.plot(x,y2,'--b',label='f2')
plt.legend(), plt.grid()
plt.title('My second Python graphs')
```

```python
def f2(x,w):
    return (x+w/2)*(x-w/2)*(x+2)

plt.figure(figsize=(15,5))
plt.subplots_adjust(wspace=0.2,hspace=0.2)
x = np.linspace(-3,3,50)
for i in range(6):
    plt.subplot(2,3,i+1);
    plt.title(i+1)
    plt.plot(x,f2(x,i+1),'k')
    plt.ylim(-20,20)
    plt.grid()
plt.show()
```

```python
def f3(x0, x1):
    r = 2*x0**2+x1**2
    return r*np.exp(-r)

n_point =30
x = np.linspace(-2,2,n_point)
y = np.linspace(-2,2,n_point)
z = np.zeros((n_point,n_point))

for m in range(len(x)):
    for n in range(len(y)):
        z[m,n] = f3(x[m],y[n])
        
```

```python
plt.figure(figsize=(3.5,3))
plt.jet()
# plt.pink(), plt.bone()
plt.pcolor(z)
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(3.5,3))
plt.jet()
c = plt.contour(x,y,z,5)
c.clabel(fmt='%3.2f', fontsize=8)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

xx, yy = np.meshgrid(x,y)

plt.figure(figsize=(5,4))

ax = plt.subplot(1,1,1,projection='3d')
ax.plot_surface(xx,yy,z,rstride=1,cstride=1,alpha=0.3,color='blue',edgecolor = 'black')
ax.set_zticks((0,0.4))
ax.view_init(45,45)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
